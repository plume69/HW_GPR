```python
from numpy import genfromtxt, zeros
# read the first 4 columns
data = genfromtxt('iris.csv',delimiter=',',usecols=(0,1,2,3))
# read the fifth column
target = genfromtxt('iris.csv',delimiter=',',usecols=(4),dtype=str)

t = zeros(len(target))
t[target == 'setosa'] = 1
t[target == 'versicolor'] = 2
t[target == 'virginica'] = 3
```

```python
import numpy as np
from numpy.random import rand
x = rand(40,1)
y = x*x*x+rand(40,1)/5
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x,y)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
gpreg = GaussianProcessRegressor()
gpreg.fit(x,y)
```

```python
from numpy import linspace, matrix
from pylab import plot, show
xx = linspace(0,1,40)
plot(x,y,'o',xx,linreg.predict(matrix(xx).T),'--r')
plot(x,y,'o',xx,gpreg.predict(matrix(xx).T),'--b')
show()
```

```python
from sklearn.metrics import mean_squared_error
print ('linreg: ',mean_squared_error(linreg.predict(x),y))
print ('gpreg:  ',mean_squared_error(gpreg.predict(x),y))
```
